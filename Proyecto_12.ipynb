## 2.5
EL modelo hace predicciones muy cercanas a los datos de validaci√≥n y la ra√≠z del error cuadratico medio nos dice que hay una variabilidad de 37.6 mil barriles.

```python
# 2.6


def entreno_validacion(df):
    X_f = df.drop(columns=['id','product'])
    y_f = df['product']
    X_trainf,X_valf,y_trainf,y_valf = train_test_split(
        X_f,y_f,test_size=.25,random_state=12345
    )
    modelf = LinearRegression()
    modelf.fit(X_trainf,y_trainf)
    
    y_pred_f = modelf.predict(X_valf)
    rmse_f = mean_squared_error(y_valf,y_pred_f,squared=False)
    mean_pred_f = y_pred_f.mean()
    mean_real_f = y_valf.mean()

    results = {
        "rmse":rmse_f,
        "mean_pred":mean_pred_f,
        "mean_real":mean_real_f,
        "y_val":y_valf.reset_index(drop=True),
        "y_pred":pd.Series(y_pred_f),
        "model":model
    }
    return results
```

```python
# 2.6
resultado_0 = entreno_validacion(data_0)
resultado_0
```

```python
# 2.6
resultado_1 = entreno_validacion(data_1)
resultado_1
```

```python
# 2.6
resultado_2 = entreno_validacion(data_2)
resultado_2
```

```python
# 3.1
region_results = {
    "region_0":resultado_0,
    "region_1":resultado_1,
    "region_2":resultado_2
}
```

```python
usd_per_unit = 4_500
top_wells = 200
budget_total = 100_000_000
cost_per_well = budget_total / top_wells
breakeven_units_per_well = cost_per_well / usd_per_unit
```

```python
def summarize_region(res:dict,name:str):
    mean_pred = res['mean_pred']
    mean_real = res['mean_real']
    rmse = res['rmse']
    return {
        "region":name,
        "rmse":rmse,
        "mean_pred":mean_pred,
        "mean_real":mean_real,
        "breakeven_units_per_well":breakeven_units_per_well,
        "pred_minus_breakeven": mean_pred - breakeven_units_per_well,
        "real_minus_breakeven": mean_real - breakeven_units_per_well,
        "meets_breakeven_pred": mean_pred >= breakeven_units_per_well,
        "meets_breakeven_real": mean_real >= breakeven_units_per_well
    }
```

```python
summary_df = pd.DataFrame(
    [summarize_region(res,name) for name,res in region_results.items()]
).set_index("region")
summary_df
```

Ninguna regi√≥n supera el umbral de equilibrio promedio ni por los datos predichos ni por los datos reales pero no se descartan las regiones todav√≠a ya que se van a evaluar los 200 mejores pozos.

```python
def top200_profit(res):
    dfv = pd.DataFrame({"y_val":res["y_val"],"y_pred":res["y_pred"]})
    top200 = dfv.nlargest(top_wells,"y_pred")
    revenue = usd_per_unit * top200["y_val"].sum()
    profit = revenue - budget_total
    summary = {
        "count_top": len(top200),
        "sum_y_val_top200":top200["y_val"].sum(),
        "sum_y_pred_top200":top200["y_pred"].sum(),
        "mean_y_val_top200":top200["y_val"].mean(),
        "mean_y_pred_top200":top200["y_pred"].mean(),
        "profit_usd":profit
    }
    return top200,summary
```

```python
results = {name:top200_profit(res) for name, res in region_results.items()}
```

```python
summary_df = pd.DataFrame({name:r[1] for name,r in results.items()}).T.sort_values("profit_usd",ascending=False)
summary_df
```

La mejor regi√≥n es la 0 ya que es la que entrega la mayor ganacia potencial ($33.21 M) y el mayor volumen promedio por pozo.

```python
n_boot = 1000
```

```python
def bootstrap(res):
    df = pd.DataFrame({"y_val": res["y_val"], "y_pred": res["y_pred"]})
    profits = []
    for _ in range(n_boot):
        sample = df.sample(500, replace=True)
        top = sample.nlargest(top_wells, "y_pred")
        revenue = usd_per_unit * top["y_val"].sum()
        profit = revenue - budget_total
        profits.append(profit)
    profits = np.array(profits)
    return {
        "mean_profit": profits.mean(),
        "ci_low": np.percentile(profits, 2.5),
        "ci_high": np.percentile(profits, 97.5),
        "risk_loss": (profits < 0).mean() * 100
    }
```

```python
results = {
    "region_0": bootstrap(resultado_0),
    "region_1": bootstrap(resultado_1),
    "region_2": bootstrap(resultado_2)
}
```

```python
for r, v in results.items():
    print(f"{r}:")
    print(f"  Ganancia media: ${v['mean_profit']:,.0f}")
    print(f"  IC95%: [${v['ci_low']:,.0f}, ${v['ci_high']:,.0f}]")
    print(f"  Riesgo de p√©rdida: {v['risk_loss']:.2f}%\n")
```

La regi√≥n 1 es la √∫nica que cuenta con un riesgo aceptable y ofrece la mayor ganancia media entre las que cumplen el umbral.
No coincide con el punto 4.3 ya que la variabilidad y los mejores promedios los tiene la regi√≥n 1 y no la 0.
